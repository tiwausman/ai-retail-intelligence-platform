```python
# Check for duplicate ZIP code prefixes

# This is more important.

geolocation["geolocation_zip_code_prefix"].duplicated().sum()
```

#### These are not duplicates—they represent different locations within the same ZIP code prefix.

### Preprocessing

If using geolocation to map customers or sellers, create a single representative coordinate for each ZIP code prefix.


```python
#Number of cities per ZIP
city_counts = (
    geolocation.groupby("geolocation_zip_code_prefix")
    ["geolocation_city"]
    .nunique()
)

city_counts.value_counts().sort_index()
```

```python
#Number of states per ZIP
state_counts = (
    geolocation
    .groupby("geolocation_zip_code_prefix")["geolocation_state"]
    .nunique()
)

state_counts.value_counts().sort_index()
```

```python
#Final aggregation 
geo_clean = (
    geolocation
    .groupby("geolocation_zip_code_prefix", as_index=False)
    .agg({
        "geolocation_lat": "mean",
        "geolocation_lng": "mean",
        "geolocation_city": lambda x: x.mode().iloc[0],
        "geolocation_state": lambda x: x.mode().iloc[0]
    })
)
```

```python
geo_clean.head()
```

The geolocation dataset contained multiple records for the same ZIP code prefix because several addresses may share a common prefix. Exact duplicate records were removed. The remaining records were aggregated to one representative location per ZIP code prefix by averaging latitude and longitude, while the most frequently occurring city and state (mode) were retained. An exploratory analysis showed that 99.96% of ZIP code prefixes mapped to a single state, indicating that the aggregation preserved the geographic integrity of the dataset.
